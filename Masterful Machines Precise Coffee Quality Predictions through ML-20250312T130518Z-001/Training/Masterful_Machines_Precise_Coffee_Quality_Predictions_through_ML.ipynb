Preprocessing

```python
df.head()
```

```python
df.nunique()
```

```python
df["Variety"].value_counts()
```

```python
df.drop(["Unnamed: 0","ID","Farm Name","Lot Number","Mill","ICO Number","Company","Altitude","Region","Producer","In-Country Partner","Grading Date",
         "Owner","Expiration","Certification Body","Certification Address","Country of Origin","Certification Contact"],axis=1,inplace = True) # The IDs, names of the companies, owner names, countries, certificates and Expiriation cont useful while predicting quality of beans.
```

```python
df.drop(["Status","Clean Cup","Sweetness","Defects"],axis = 1,inplace = True) # these columns has only one unique value in each column, we can predict quality using these.
```

```python
def segment_var(Variety):
    if pd.isnull(Variety):
        return 'Other'
    elif "Caturra" in Variety:
        return 'Caturra'
    elif 'Gesha' in Variety:
        return 'Gesha'
    elif 'Typica' in Variety:
        return 'Typica'
    elif 'Bourbon' in Variety:
        return 'Bourbon'
    elif 'Catuai' in Variety:
        return 'Catuai'
    elif 'unknown' in Variety:
        return 'unknown'
    elif 'Catimor' in Variety:
        return 'Catimor'
    elif "Ethiopian Heirlooms" in Variety:
        return "Ethiopian"
    elif "SL34" in  Variety:
        return "SL34"  
    elif "Mundo Novo" in  Variety:
        return "Mundo Novo"     
    elif "SL14" in  Variety:
        return "SL14" 
    elif "Yellow Bourbon" in  Variety:
        return "Yellow Bourbon" 
    elif "SHG" in  Variety:
        return "SHG" 
    elif "Java" in  Variety:
        return "Java"             
    else:
        return 'Other'

df['Variety'] = df['Variety'].apply(segment_var) #Extracting important labels from Variety column and rest of them which are having one-time repeated values converted into others
```

```python
from sklearn.impute import SimpleImputer

# Create a SimpleImputer instance with the strategy as "most_frequent"
imputer = SimpleImputer(strategy='most_frequent')

# Fill missing values in categorical columns
df[cat_cols] = imputer.fit_transform(df[cat_cols])
```

```python
x = df.drop("Total Cup Points",axis=1)
y = df["Total Cup Points"]
# specifing target column and independent columns and assigning t0 y and x
```

```python
from sklearn.model_selection import train_test_split
print("#Spliting data into 80:20 as train and test")
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) 
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
y_train
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x_train["Harvest Year"] = le.fit_transform(x_train["Harvest Year"])
x_test["Harvest Year"] = le.transform(x_test["Harvest Year"])

x_train["Variety"] = le.fit_transform(x_train["Variety"])
x_test["Variety"] = le.transform(x_test["Variety"])

x_train["Processing Method"] = le.fit_transform(x_train["Processing Method"])
x_test["Processing Method"] = le.transform(x_test["Processing Method"])

x_train["Color"] = le.fit_transform(x_train["Color"])
x_test["Color"] = le.transform(x_test["Color"])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.transform(x_test)
```

```python
x_train_std
```

Model building

```python
from sklearn.linear_model import LinearRegression

# Initialize the linear regression model
lr = LinearRegression()

# Fit the model to the training data
lr.fit(x_train_std, y_train)
y_train_pred = lr.predict(x_train_std)
y_test_pred = lr.predict(x_test_std)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
scores = pd.DataFrame(columns=["model", "train_mse","train_rmse","train_mae","train_r2","test_mse","test_rmse","test_mae","test_r2"])
def Metrics(y_train , y_train_pred , y_test , y_test_pred,model,dataframe):
  # calculate the mean squared error (MSE)
  train_mse = mean_squared_error(y_train,y_train_pred)

  # calculate the root mean squared error (RMSE)
  train_rmse = np.sqrt(train_mse)

  # calculate the mean absolute error (MAE)
  train_mae = mean_absolute_error(y_train,y_train_pred)

  # calculate the coefficient of determination (R-squared)
  train_r2 = r2_score(y_train,y_train_pred)

   # calculate the mean squared error (MSE)
  test_mse = mean_squared_error(y_test,y_test_pred)

  # calculate the root mean squared error (RMSE)
  test_rmse = np.sqrt(test_mse)

  # calculate the mean absolute error (MAE)
  test_mae = mean_absolute_error(y_test,y_test_pred)

  # calculate the coefficient of determination (R-squared)
  test_r2 = r2_score(y_test,y_test_pred)
  dataframe = dataframe.append(pd.Series([model, train_mse,train_rmse,train_mae,train_r2,test_mse,test_rmse,test_mae,test_r2],index=scores.columns ), ignore_index=True)
# print the evaluation metrics
  return(dataframe)
```

```python
scores = Metrics(y_train , y_train_pred ,y_test , y_test_pred , "linear regression",scores)
scores
```

```python
from sklearn.tree import DecisionTreeRegressor 
  
# create a regressor object
DT_reg = DecisionTreeRegressor(max_depth=5, min_samples_split=5,random_state = 0) 
  
# fit the regressor with X and Y data
DT_reg.fit(x_train_std, y_train)
```

```python
y_train_pred = DT_reg.predict(x_train_std)
y_test_pred=DT_reg.predict(x_test_std) 
```

```python
scores = Metrics(y_train , y_train_pred ,y_test , y_test_pred , "DT",scores)
scores
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(n_estimators=100, random_state=42)
rf_reg.fit(x_train_std, y_train)
```

```python
y_train_pred = rf_reg.predict(x_train_std)
y_test_pred = rf_reg.predict(x_test_std)
```

```python
scores = Metrics(y_train , y_train_pred ,y_test , y_test_pred , "RandomForestReg",scores)
scores
```

```python
import xgboost as xgb
xgb_reg = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
xgb_reg.fit(x_train_std, y_train)
y_pred_train = xgb_reg.predict(x_train_std)
y_test_pred = xgb_reg.predict(x_test_std)
```

```python
scores = Metrics(y_train , y_train_pred ,y_test , y_test_pred , "XGB",scores)
scores
```
